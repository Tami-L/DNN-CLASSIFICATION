```python
data = pd.read_csv('/Users/lindokuhletami/Downloads/diabetes.csv')
```

```python
data.head()
```

```python
data.replace('?', None, inplace = True)
```

```python
data['diag_1'] = data['diag_1'].str.extract(r'(\d+.\d+)')
data['diag_2'] = data['diag_2'].str.extract(r'(\d+.\d+)')
data['diag_3'] = data['diag_3'].str.extract(r'(\d+.\d+)')
```

```python
data.isnull().sum()/len(data)*100
```

```python
data.columns
```

```python
data.drop(['id', 'encounter_id', 'patient_nbr','payer_code', 'medical_specialty','weight','max_glu_serum', 'A1Cresult'], axis = 1,inplace = True)
```

```python
data['diabetesMed'].value_counts()
```

```python
yes_class = data[data['diabetesMed']== 'Yes']
no_class = data[data['diabetesMed']== 'No']
```

```python
df = yes_class.sample(n= 23403, replace=False)

balanced_data = pd.concat([df,no_class], axis = 0 )
```

```python
balanced_data.isnull().sum()/len(balanced_data)*100
```

```python
balanced_data['race'].value_counts()
```

```python
import numpy as np
balanced_data['race']= balanced_data['race'].replace(np.nan,'Caucasian')
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')

balanced_data['diag_1']= imputer.fit_transform(balanced_data['diag_1'].values.reshape(-1,1))
balanced_data['diag_2']= imputer.fit_transform(balanced_data['diag_2'].values.reshape(-1,1))
balanced_data['diag_3']= imputer.fit_transform(balanced_data['diag_3'].values.reshape(-1,1))
```

# prep for model

```python
cat = pd.get_dummies(balanced_data.select_dtypes(exclude='number'), drop_first= True,dtype= 'int')
cat
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
col_names =balanced_data.select_dtypes(include='number').columns
num = pd.DataFrame(scale.fit_transform(balanced_data.select_dtypes(include='number')),columns = col_names)
num
```

```python
num.reset_index(drop = True, inplace = True)
cat.reset_index(drop = True, inplace = True)

data = pd.concat([cat,num], axis =1)
data
```

```python
y = data['diabetesMed_Yes']
X = data.drop(['diabetesMed_Yes'], axis = 1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.heatmap(X.corr())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(X,y,test_size= 0.3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
len(X.columns)
```

```python
model = Sequential()
model.add(Dense(74, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(40, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation = 'relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer= 'adam', loss = 'binary_crossentropy')
```

```python
model.fit(x_train,y_train,epochs = 20, validation_data = (x_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
predictions = [int(x) for x  in model.predict(x_test)]
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
np.log(losses['loss']).plot(title = 'log loss', xlabel = 'epoch',ylabel = 'loss' )
```
